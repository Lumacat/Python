```python
# Rosalind / PMCH

import math
import networkx as nx
import random
from Bio import SeqIO

with open('/Users/Shared/Rosalind/rosalind_pmch.txt', 'r') as my_file:
    lst = [x.seq for x in SeqIO.parse(my_file, 'fasta')]    
    s = lst[0]

    print(math.factorial(s.count('A')) * (math.factorial(s.count('C'))))
    
    ls_A, ls_C, ls_G, ls_U, ls_tot = [], [], [], [], []
    A_count, C_count, G_count, U_count = 0, 0, 0, 0

    for x in list(str(s)):
        if x == 'A':
            A_count += 1
            ls_A.append(x + str(A_count))
        if x == 'C':
            C_count += 1
            ls_C.append(x + str(C_count))
        if x == 'G':
            G_count += 1
            ls_G.append(x + str(G_count))
        if x == 'U':
            U_count += 1
            ls_U.append(x + str(U_count))
            
    G = nx.Graph()
    ls_A2, ls_C2, ls_G2, ls_U2 = ls_A.copy(), ls_C.copy(), ls_G.copy(), ls_U.copy()
    
    for x in str(s):
        if x == 'A':
            ls_tot.append(ls_A[0])
            ls_A.remove(ls_A[0])
        if x == 'C':
            ls_tot.append(ls_C[0])
            ls_C.remove(ls_C[0])
        if x == 'G':
            ls_tot.append(ls_G[0])
            ls_G.remove(ls_G[0])
        if x == 'U':
            ls_tot.append(ls_U[0])
            ls_U.remove(ls_U[0])
            
    for x in ls_tot:
        G.add_node(x)
        
    random.shuffle(ls_A2)
    random.shuffle(ls_C2)
    random.shuffle(ls_G2)
    random.shuffle(ls_U2)
    
    for i in range(len(ls_A2)):
        G.add_edge(ls_A2[i],ls_U2[i])
    for i in range(len(ls_C2)):
        G.add_edge(ls_C2[i],ls_G2[i])
    
    pos = nx.circular_layout(G)
    nx.draw(G, pos = pos, with_labels=True)
```
