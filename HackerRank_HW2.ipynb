```python
# Algorithms / Warmup / Compare the triplets

#!/bin/python3

import math
import os
import random
import re
import sys

def compareTriplets(a, b):
    # Write your code here
    scores = [0, 0]
    for i in range(len(a)):
        if a[i] > b[i]:
            scores[0] += 1
        elif a[i] < b[i]:
            scores[1] += 1
        else:
            pass
    return(scores)

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    a = list(map(int, input().rstrip().split()))

    b = list(map(int, input().rstrip().split()))

    result = compareTriplets(a, b)

    fptr.write(' '.join(map(str, result)))
    fptr.write('\n')

    fptr.close()
```

```python
# Algorithms / Warmup / Diagonal difference

#!/bin/python3

import math
import os
import random
import re
import sys

def diagonalDifference(arr):
    # Write your code here
    d1_s = []
    d2_s = []
    for i in range(len(arr)):
        d1_s.append(arr[i][i])
        d2_s.append(arr[i][int(len(arr))-1-i])
    return(abs(sum(d1_s)-sum(d2_s)))

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    n = int(input().strip())

    arr = []

    for _ in range(n):
        arr.append(list(map(int, input().rstrip().split())))

    result = diagonalDifference(arr)

    fptr.write(str(result) + '\n')

    fptr.close()
```

```python
# Algorithms / Warmup / Min-Max-Sum

#!/bin/python3

import math
import os
import random
import re
import sys

def miniMaxSum(arr):
    t_max = 0
    t_min = sum(arr)

    for i in range(len(arr)):
        t_max_temp = sum(arr)-arr[i]
        if t_max_temp > t_max:
            t_max = t_max_temp
    
    for i in range(len(arr)):
        t_min_temp = sum(arr)-arr[i]
        if t_min_temp < t_min:
            t_min = t_min_temp

    print(t_min, t_max)

if __name__ == '__main__':

    arr = list(map(int, input().rstrip().split()))

    miniMaxSum(arr)
```

```python
# Algorithms / Warmup / Time conversion

#!/bin/python3

import math
import os
import random
import re
import sys

def timeConversion(s):
    # Write your code here
    x = s.split(":")
    if x[2][2] == 'P':
        if x[0] != '12':
            x[0] = str(int(x[0])+12)
        x[2] = x[2][:-2]
    else:
        if x[0] == '12':
            x[0] = '00'
        x[2] = x[2][:-2]
    return(x[0] + ':' + x[1] + ':' + x[2])
    
if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    s = input()

    result = timeConversion(s)

    fptr.write(result + '\n')

    fptr.close()
```

```python
# Algorithms / Sorting / Correctedness of the loop invariant

def insertion_sort(l):
    for i in range(1, len(l)):
        j = i-1
        key = l[i]
        while (j >= 0) and (l[j] > key): # j has to be set > or = 0
           l[j+1] = l[j]
           j -= 1
        l[j+1] = key

m = int(input().strip())
ar = [int(i) for i in input().strip().split()]
insertion_sort(ar)
print(" ".join(map(str,ar)))
```

```python
# Algorithms / Sorting / Running time of algorithms

#!/bin/python3

import math
import os
import random
import re
import sys

def runningTime(arr):
    count = 0
    for i in range(1,len(arr)):
        for j in range(0,i):
            if arr[i] < arr[j]:
                arr.insert(j,(arr.pop(i)))
                count += i-j
    return(count)

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    n = int(input().strip())

    arr = list(map(int, input().rstrip().split()))

    result = runningTime(arr)

    fptr.write(str(result) + '\n')

    fptr.close()
```

```python
# Algorithms / Sorting / Quicksort 1 - Partition

#!/bin/python3

import math
import os
import random
import re
import sys

def quickSort(arr):
    center = []
    left = []
    right = []
    center.append(arr[0])
    
    for i in range(1,len(arr)):
        if arr[i] > arr[0]:
            right.append(arr[i])
        else:
            left.append(arr[i])
    m = left + center + right
    return(m)

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    n = int(input().strip())

    arr = list(map(int, input().rstrip().split()))

    result = quickSort(arr)

    fptr.write(' '.join(map(str, result)))
    fptr.write('\n')

    fptr.close()
```

```python
# Algorithms / Strings / CamelCase

#!/bin/python3

import math
import os
import random
import re
import sys

def camelcase(s):
    words = 1
    for x in s:
        if x.isupper() == True:
            words += 1
    return(words)

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    s = input()

    result = camelcase(s)

    fptr.write(str(result) + '\n')

    fptr.close()
```

```python
# Algorithms / Strings / Super reduced string

#!/bin/python3

import math
import os
import random
import re
import sys

def superReducedString(s):
    l=[]
    for x in s:
        l.append(x)
    count = 1
    while count < len(l):
        if l[count] == l[count-1]:
            l.pop(count)
            l.pop(count-1)
            count = 1
        else:
            count +=1
    if l == []:
        return('Empty String')
    else:
        return(''.join(l))

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    s = input()

    result = superReducedString(s)

    fptr.write(result + '\n')

    fptr.close()
```
